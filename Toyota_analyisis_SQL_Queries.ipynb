# q1_avg_price_by_model_top10

```python
%sql
SELECT
  model,
  ROUND(AVG(price), 2) AS avg_price,
  COUNT(*) AS num_cars
FROM workspace.toyota_project.toyota_clean
GROUP BY model
HAVING COUNT(*) >= 10
ORDER BY avg_price DESC
LIMIT 10;
```

#q2_price_by_fuel_and_transmission

```python
%sql
SELECT
  fuelType,
  transmission,
  ROUND(AVG(price), 2) AS avg_price,
  COUNT(*) AS num_cars
FROM workspace.toyota_project.toyota_clean
GROUP BY fuelType, transmission
ORDER BY avg_price DESC;
```

#q3_price_vs_car_age


```python
%sql
SELECT
  car_age,
  ROUND(AVG(price), 2) AS avg_price,
  ROUND(AVG(depreciation_score), 2) AS avg_depreciation_score
FROM workspace.toyota_project.toyota_clean
GROUP BY car_age
ORDER BY car_age;
```

# q4_price_by_mileage_band

```python
%sql
SELECT
  mileage_band,
  ROUND(AVG(price), 2) AS avg_price,
  COUNT(*) AS num_cars
FROM workspace.toyota_project.toyota_clean
GROUP BY mileage_band
ORDER BY avg_price DESC;
```

# q5_engine_size_vs_efficiency

```python
%sql
SELECT
  engineSize,
  ROUND(AVG(fuel_efficiency_index), 3) AS avg_fuel_efficiency,
  ROUND(AVG(mpg), 2) AS avg_mpg,
  COUNT(*) AS num_cars
FROM workspace.toyota_project.toyota_clean
GROUP BY engineSize
ORDER BY engineSize;
```
